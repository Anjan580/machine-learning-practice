```python
import seaborn as sns
from sklearn.model_selection import train_test_split
import pandas as pd

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('Diabetes(in).csv')
df.head()
```

```python
# Scaling is done after train_test split
```

```python
x = df[['Glucose','Age']]
y = df.Diabetes
```

```python
x.head()
```

```python
# train- test split
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size = 0.2, random_state = 5)
```

y doesn't need scaling

```python
# after train_test split
# x : xtrain , xtest
# xtrain : .fit_transform()
# xtest : .transform()

scaler = StandardScaler()
xtrain_s = scaler.fit_transform(xtrain)

xtest_s = scaler.transform(xtest)
```

```python
# scaling
scaler = StandardScaler()
xscaled = scaler.fit_transform(x)
```

```python
xscaled
```

```python
knn = KNeighborsClassifier()
knn.fit(xtrain_s, ytrain)
```

```python
# Evaluation 
yprediction = knn.predict(xtest_s)
```

```python
accuracy = accuracy_score(ytest, yprediction)
print(f'Accuracy : {accuracy:.2%}')
```

# Dicision Tree

```python
import seaborn as sns
```

```python
data = sns.load_dataset('iris')
data.head()
```

```python
x =data.drop('species', axis= 1)
y = data.species
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.15, random_state = 12)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dmodel = DecisionTreeClassifier()
dmodel.fit(xtrain, ytrain)
```

```python
# prediction 
# Evaluation
# view tree
```

```python
# prediction
dmodel.predict([[5.5, 2.5, 1,1]])
```

```python
dmodel.predict([[5.0, 3.6, 1.4,	0.2]])
```

```python
# Evaluation
yprediction = dmodel.predict(xtest)
```

```python
yprediction
```

```python
accuracy = accuracy_score(ytest, yprediction)
print(f'Accuracy:{accuracy: .2%}')
```

```python
# view tree
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize = (10,10))
plot_tree(dmodel, feature_names = x.columns, class_names = y.unique(), filled = True)
plt.show()
```
