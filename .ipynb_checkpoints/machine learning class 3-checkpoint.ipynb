# multiple linear regression

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error, r2_score
```

```python
df = pd.read_csv('salary_Data.csv')
df.head()
```

```python
df.dropna(inplace = True)
```

```python
# to change the column name
df = df.rename(columns = {
    'Years of Experience': 'yoe'
})
```

```python
df.columns
```

```python
df.head()
```

```python
x = df[['Age', 'yoe']]
y= df['Salary']
```

```python
x
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size= 0.2, random_state = 15)
```

```python
ml_model = LinearRegression()
ml_model.fit(xtrain, ytrain)
```

```python
ml_model.predict([[40,5]])
```

```python
# Evaluation
yprediction = ml_model.predict(xtest)
```

```python
mes = mean_squared_error(ytest, yprediction)
print(mes)
```

```python
mae = mean_absolute_error(ytest, yprediction)
print(mae)
```

```python
mape = mean_absolute_percentage_error(ytest, yprediction)
print(mape)
```

```python
r2 = r2_score(ytest, yprediction)
print(r2)
```

# Logistic Regression

Used for classification

```python
df = pd.read_csv('Diabetes(in).csv')
df.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
x = df[['Age']]
y = df['Diabetes']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size = 0.2, random_state = 17)
```

```python
log_model = LogisticRegression()
log_model.fit(xtrain, ytrain)
```

```python
# age  = 50
print(log_model.predict([[50]]))
```

```python
print(log_model.predict_proba([[50]]))
```
