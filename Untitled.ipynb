```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('Mall_Customers.csv')
df.head()
```

```python
df.columns
```

```python
df = df.rename(columns = {
     'Annual Income (k$)': 'Annual_income',
    'Spending Score (1-100)':'Spend_score'
    
})
```

```python
df.head()
```

```python
df_n = df[['Annual_income', 'Spend_score']]
df_n.head()
```

```python
plt.scatter(df_n.Annual_income, df_n.Spend_score)

plt.xlabel('Annual income')
plt.ylabel('Spending score')
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
x = df_n
```

```python
# Scaling
scaler  = StandardScaler()
xscaled = scaler.fit_transform(x)
```

```python
kmean = KMeans(n_clusters = 5)
kmean.fit(xscaled)
```

```python
label = kmean.labels_
label
```

```python
plt.scatter(df_n.Annual_income, df_n.Spend_score, c = label)
plt.show()
```

```python
df_n['cluster'] = label
df_n.head()
```

```python
kmean.predict([[-1, -0.1]])
```
